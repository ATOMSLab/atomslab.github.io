```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
import plotly.figure_factory as ff
import plotly.graph_objects as go
```

```python
#All lab members
df = pd.read_excel('/Users/rediaz/GitHub/fraser-lab.github.io/members_timeline/fraser_timeline.xlsx')
# Gantt chart in Plotly has mandatory format of data columns
df.columns = ['Task', 'Start', 'Finish', 'Role']

fig_plotly = ff.create_gantt(df, index_col='Role', title='Fraser Lab 2011-2021',
                             show_colorbar=True, bar_width=0.4, showgrid_x=True, showgrid_y=True, group_tasks=True)

fig_plotly.update_layout(
    autosize=True,
    width=1000,
    height=1000,
    margin=go.layout.Margin(
        l=150,
        r=5,
        b=100,
        t=100,
        pad=4
    )
)
```
